```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_columns = 50
pd.options.display.max_rows = 250
```

```python
df = pd.read_csv('../Datasets/CarPrice_Assignment.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns = [cols.lower() for cols in df.columns]
```

```python
df
```

```python
df['brand'] = df['carname'].str.extract(r'(\w+)')
```

```python
df[['carname','brand']]
```

```python
df['brand'] = df['brand'].replace({'vw':'volkswagen'})
df['brand'] = df['brand'].replace({'vokswagen':'volkswagen'})
df['brand'] = df['brand'].replace({'toyouta':'toyota'})
df['brand'] = df['brand'].replace({'porcshce':'porsche'})
df['brand'] = df['brand'].replace({'maxda':'mazda'})
df['brand'] = df['brand'].replace({'Nissan':'nissan'})
```

```python
df['brand'].unique().tolist()
```

### One Hot Encoding

```python
car_brand = pd.get_dummies(df['brand'])
fuel_type = pd.get_dummies(df['fueltype'])
aspiration = pd.get_dummies(df['aspiration'])
car_type = pd.get_dummies(df['carbody'])
fuel_system = pd.get_dummies(df['fuelsystem'])
drive_wheel = pd.get_dummies(df['drivewheel'])
```

```python
df.columns
```

```python
df.head(3)
```

```python
df['cylindernumber'].value_counts()
```

```python
num_col = df[['carlength','carwidth','carheight','wheelbase',
         'curbweight','enginesize','horsepower','citympg','highwaympg','peakrpm']]
num_col
```

### Using Standard Scaler

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(num_col)
scaled_data = pd.DataFrame(scaled_data,columns=num_col.columns)
scaled_data
```

```python
car_brand = pd.DataFrame(car_brand)
fuel_type = pd.DataFrame(fuel_type)
aspiration = pd.DataFrame(aspiration)
car_type = pd.DataFrame(car_type)
fuel_system = pd.DataFrame(fuel_system)
drive_wheel = pd.DataFrame(drive_wheel)
```

```python
df2 = pd.concat([car_brand,fuel_type,aspiration,car_type,fuel_system,drive_wheel],axis=1)
df2
```

```python
df3 = pd.concat([df2,scaled_data],axis=1)
df3['price'] = df['price']
df3
```

### Splitting Features and Target

```python
X = df3.drop('price',axis=1)
Y = df3['price']
```

```python
X.head()
```

```python
Y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=1,shuffle=True)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print("Mean Absolute Error :",mean_absolute_error(y_test,predict))
print("Mean Squared Error :",mean_squared_error(y_test,predict))
print("R2 Score :",r2_score(y_test,predict)*100)
```
